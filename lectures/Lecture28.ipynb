# List (Continue)--List Comprehensions


Python's list comprehensions let you create lists in a way that is reminiscent of set notation

$$ S = \{ x^2 ~\mid~ 0 \le x \le 20, x\mod 3 = 0\}$$

````python
S = []
for x in range(21):
    if x % 3 == 0:
        S.append(x**2)
S
````

Syntax is generally
```python3
S = [<element> <for statement> <conditional>]
```

````python
S = [x**2 for x in range(21) if x % 3 == 0]
S
````

Nested for loop

````python
S = []
for i in range(2):
    for j in range(2):
        for k in range(2):
            if (i + j + k) % 2 == 0:
                S += [[i, j, k]]

S
````

````python
# you aren't restricted to a single for loop
S = [
    [i, j, k]                  #<element> <for statement> <conditional>
    for i in range(2)          #<for statement>
    for j in range(2)
    for k in range(2)
    if (i + j + k) % 2 == 0    #<conditional>
]
S
````

````python
# you aren't restricted to a single if statement : Ternary operation
# s1 if c1 else c2
["yes" if v == 1 else "no" if v == 2 else "idle" for v in [1, 2, 3]]
````

## Exercise 1: Class Cat

````python
class Cat:  # declare a class named Cat
    # constructor __init__
    # __init__ method initializes an instance of the Cat class, it runs automatically when an object is created
    # the parameter input name is optional, with a default value of None
    def __init__(self, name = None):
        # self = the current object
        # the argument name is assigned to the instance variable self.name
        self.name = name
        # a message is printed to indicate the creation of a Cat object with the provided name
        print(f"A Cat class with name {name} is created :-)")
    # method
    def set_age(self, age = None):
        # update the age of the instance/object
        # assign the new age argument to self.age
        self.age = age
    def state_age(self):
        # print the current value of self.age
        print(f"{self.name} is {self.age} years old")
    def cat_hello(self, sound:str = 'Meow'):
        print(f"{self.name} says: {sound}!")
````

````python
Frisky = Cat(name="Frisky")
````

````python
Frisky.cat_hello()
````

````python
Frisky.set_age(age = 3)
Frisky.state_age()
````

````python
Betty = Cat(name = "Betty")
Betty.set_age(age = 5)
Betty.state_age()
Betty.cat_hello(sound = "Purr")
````
