```python
%%timeit
# This is a magic command available in IPython environments like Jupyter Notebook.
# It automatically measures the execution time of the entire cell.
# It runs the code multiple times to compute an average execution time for better accuracy.
x = list(np.random.rand(10000))
y = list(np.random.rand(10000))
z = 0
for i in range(10000):
    z += x[i] * y[i]
```
