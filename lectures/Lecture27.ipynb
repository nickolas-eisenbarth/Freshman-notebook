````python
def triangle_area(base, height):
    try:
        if base < 0 or height < 0:
            raise ValueError("Base and height must be non-negative")
        return 0.5 * base * height
    except ValueError as e_message: # capture the error in e_message and print custom error message
        print(f"Error, triangle_area {e_message}, Try a different non-negative value")
    except Exception as e: # handle any other unexpected error that is not a ValueError
        print(f"Error,triangle_area new error {e}")


triangle_area(1, 2)
````

````python
# everything in python is an object, and can be passed into a function
def f(x):
    return x + 2


def twice(g, x):
    return g(g(x))


twice(f, 2)  # + 4
````

````python
def n_apply(f, x, n):
    """applies f to x n times"""
    for _ in range(n):  # _ is dummy variable in iteration
        x = f(x)
    return x


n_apply(f, 1, 5)  # 1 + 2*5
````

How a.copy() works:


1.   creates a new list object and the new list is independent of the original list
2.   **shallow copy**: copy references for nested objects: for elements that are themselves objects (e.g., sublists or other mutable objects), the copy contains references to the same objects, not duplicates.


````python
a = [1, 2, [3, 4]]
b = a.copy()# shallow copy

# Modify the inner list in the copied list b
b[2][0] = 99

print("Original List a:", a)
print("Copied List b:", b)
````

````python
a = [1, 2, [3, 4]]
import copy
b = copy.deepcopy(a)# deep copy

# Modify the inner list in the copied list b
b[2][0] = 99

print("Original List a:", a)
print("Copied List b:", b)
````

````python
?copy.deepcopy
````

````python
a = [1, 2, [3, [4,5]]]
import copy
b = copy.deepcopy(a)# deep copy

# Modify the inner list in the copied list b
b[2][1][1] = 99

print("Original List a:", a)
print("Copied List b:", b)
````

Nested for loop

````python
S = []
for i in range(2):
    for j in range(2):
        for k in range(2):
            if (i + j + k) % 2 == 0:
                S += [[i, j, k]]

S
````

````python
# you aren't restricted to a single for loop
S = [
    [i, j, k]                  #<element> <for statement> <conditional>
    for i in range(2)          #<for statement>
    for j in range(2)
    for k in range(2)
    if (i + j + k) % 2 == 0    #<conditional>
]
S
````

* We've seen the `list` class, which is ordered/indexed, and mutable. `[ ]`

There are other Python collections that you may find useful:
* `tuple` which is ordered/indexed, and immutable. `( )`
* `set` which is unordered/unindexed, mutable, and doesn't allow for duplicate elements. `{ }`
* `dict` (dictionary), paired, which is unordered/unindexed, and mutable, with no duplicate keys. `{key0: value0, key1: value1}`

````python
a_tuple = (1, 2, 4) # notice the difference with list a_list = [1,2,4]
print(a_tuple)
a_tuple[0] = 1 # immutable
````

````python
a_set = {5, 3, 2, 5.0}# notice the difference with list a_list = [1,2,4], a_tuple = (1,2,4)
print(a_set) # no duplicate element
print(a_set[0]) # no index
````

````python
a_set.add(6)  # you can also add all element in a list by using .update()
print(f"After adding 6,   a_set:{a_set}")
a_set.remove(6)  # you can also add all element in a list by using minus operator
print(f"After removing 6, a_set:{a_set}")
````
