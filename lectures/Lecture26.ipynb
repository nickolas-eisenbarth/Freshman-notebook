# 1. Control Flows


**if statements:**
if decides whether certain statements need to be executed or not. It checks for a given condition.

```python
if condition1:
  statements1
elif condition2:
  statements2
elif condition3:
  statements3
else:
  statements4
```

````python
x = 10
y = 4

if x == y:
    print("The value of x is the same as value of y")
elif x == 3:
    print("I am here")
    print("x is 3")
else:
    print("x is something else")
````

**For loops**: control flow statement for specifying iteration, which allows code to be executed repeatedly.

````python
for i in range(4):  # default - start at 0, increment by 1
    print(f"{i}, next loop")
````

````python
for i in range(10, 2, -2):  # inputs are (start, stop, step)
    print(f"{i}")
    print(i)
````

````python
fruits = ["apple", "banana", "cherry"] #iteration through a list
for fruit in fruits:
    print(f"I like {fruit}")
````

**while loop**: Executes as long as the condition is `True`.

````python
# Find all positive integer whose square < 100
i = 1
while i**2 < 100:
    print(i)
    i += 1  # i += 1 is the same as i = i + 1
print("Finished")
````

**continue:** skip the rest of the loop

**break:** exit from the loop

````python
# distinguish even or odd number
for num in range(1, 10):
    if num % 2 == 0:
        print(f"Found {num}, an even number, Continue!")
          # this jumps us back to the top

    print(f"Found {num}, an odd number")


print("Finished")
````

````python
for num in range(1, 10):
    if num % 2 == 0:  # if n divisible by x
        print(f"Found {num}, an even number, break!")
        break

    print(f"Found {num}, an odd number")
print("Finished")
````

````python
for num in range(1, 10):
    if num % 2 == 0:  # if n divisible by x
        print(f"Found {num}, an even number, Continue!")
        continue

    print(f"Found {num}, an odd number")
print("Finished")
````

````python
# continue/break statement only continue/breaks its closest parent level loop statement
for i in range(3):
    print(f"Outer {i}")
    for j in range(3):
        print(f"   Inner : {j}")
        break
    # jump to here after break
````

**try, except, finally**: used for handling exceptions.

````python
100/0
````

````python
try:
    print(100 / 0)
except ZeroDivisionError:
    print("Error: don't divided by zero")
except NameError:
    # print("Error: the variabel is not defined")
    pass
except Exception as e_msg:# a general exception handler, it catches any error that hasn’t been specifically handled by the earlier except blocks.
    print("we have an exception")
    print(f"Other Error: {e_msg}")
finally:# The finally block always executes, regardless of whether an error occurred or not.
  print("Execution completed.")
````

**match and case** (Python 3.10+): used for pattern matching.

````python
command = "start"

match command:
    case "start":
        print("Starting the process.")
    case "stop":
        print("Stopping the process.")
    case _:
        print("Unknown command.")
````

````python
def triangle_area(base, height):
    if base < 0 or height < 0:
        raise ValueError("Base and height must be non-negative")
    return 0.5 * base * height


triangle_area(-1, 2)
````
