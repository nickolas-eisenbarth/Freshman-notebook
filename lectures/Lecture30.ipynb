### Exercise

Create a numpy array that contain  intergers $i$  such that  $0<i<100$ and $2^i$ has the last digit 6

```python
l = [i for i in range(1, 100) if 2**i // 10 == 6]
np.array(l)
```

Create a 2D numpy array $A$ (5,10) such that $A_{ij} = i\times j$

```python
# Method 1 (inefficient): initialize an empty array and assign values
A = np.zeros(shape=(5, 10))
for i in range(5):
    for j in range(10):
        A[i, j] = i * j
A
```

```python
# Method 2 (efficient): initialize an nested list and transform it to an array
l = [[i * j for j in range(10)] for i in range(5)]
np.array(l)
```

#2. Numpy Array Dimension and Reshape
if `X` is an numpy array,
- You can check the dimension/shape of `X` by `X.shape` or `np.shape(X)`
- You can reshape `X` to any possible shape by `np.reshape(X, shape)`

```python
print(np.arange(3))
print(np.arange(3).shape)
```

```python
array1 = np.arange(3) * np.arange(3) #element wise multi, shape (3,)
print(array1)
print(array1.shape)
array1 = array1.reshape(1,-1) #reshape (1,3)
print(array1.shape)
print(array1)
```

```python
array2 = np.arange(3).reshape(3, 1)
print(array2.shape)
print(array2)
```

```python
# This is a 1D array
Z = np.zeros(shape=5)
print(Z)
Z.shape
```

```python
Z = np.zeros(shape = (5,5))
print(Z)
Z.shape
```

```python
# Create 2D array by reshape
Z = np.zeros(5).reshape(5, 1)
print(Z)
Z.shape
```

```python
Z.squeeze()  # remove axis with length = 1
```

```python
(Z.squeeze()).shape
```

```python
# squeeze Remove axes of length one
Z = np.zeros(6).reshape(1, 1, 2, 3)
print(Z)
print(Z.shape, "\n")

Z_squeeze = Z.squeeze()
print(Z_squeeze)
print(Z_squeeze.shape)
```

```python
A = np.arange(15).reshape(5,3)
print(A.shape)
A
```

# 3. Array Broadcasting


Normally you only do arithmetic operations between arrays of the same dimension. But it is also possible to do operations between arrays of different size (not always). When it is possible, numpy array is doing **array braodcasting**.

The smaller array of at least 1 dimension of size 1 is “broadcast” across the larger array so that they have compatible shapes by dimension.

It tries to duplicates every x’s col and y’s row (dim size = 1) to match the other arrays
1. Make the two arrays have the same number of dimensions.
  - If the numbers of dimensions of the two arrays are different, add new dimensions with size 1 to the head of the array with the smaller dimension.

2. If there is a dimension whose size is not 1 in either of the two arrays, it cannot be broadcasted, and an error is raised.

![](http://scipy-lectures.org/_images/numpy_broadcasting.png)

```python
X = np.arange(4).reshape(-1, 1) * 10
Y = np.arange(3).reshape(1, -1)
print(X.shape)
print(Y.shape, "\n")
print(X, "\n")
print(Y, "\n")
X + Y
```

```python
a = np.arange(3).reshape(1, 3)
c = np.ones((3, 3))
d = np.zeros((6, 3))
print(a, "\n")
print(c, "\n")
print(d, "\n")
```

```python
print(c)
print(c + 2)
```

```python
print(d)
print(a)
print(d + a)
```
